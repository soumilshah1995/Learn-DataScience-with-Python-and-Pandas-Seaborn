# how may job title begin with CAPTAIN

```python
def check(x):
    if 'chief' in x.lower().split():
        return True
    else:
        return False
```

```python
df[df["JobTitle"].apply(lambda x : check(x))]
```

```python
df["JobTitle"].apply(lambda x : check(x)).value_counts()
```
